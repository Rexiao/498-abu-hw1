# Q1

```python
dfs = []
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        dfs.append(pd.read_csv(os.path.join(dirname, filename),index_col=0))
```

```python
nlikes = 0
tweets_num = 0
for df in dfs:
    nlikes += df['nlikes'].sum()
    tweets_num += df['nlikes'].size
like_avg = nlikes / tweets_num
    
```

```python
print(like_avg)
print(round(like_avg))
```

# Q2

```python
nreplies = 0
nretweets = 0
for df in dfs:
    nreplies += df['nreplies'].sum()
    nretweets += df['nretweets'].sum()
```

```python
engagement_rate = 0.05
views_total = (nlikes+nreplies+nretweets) / engagement_rate
views_avg = views_total / tweets_num
```

```python
print(views_avg)
print(round(views_avg))
```

# Q3

```python
types = dfs[0].dtypes
```

```python
types
```

```python
i64_count = types[types == np.int64].size
f64_count = types[types == np.float64].size
```

```python
strlen_total = 0
indices = items = types[types==np.object].index
for df in dfs:
    strlen_total += df.filter(indices).applymap(lambda x: len(str(x))).to_numpy().sum()
    
strlen_avg = strlen_total / tweets_num
```

```python
video_avg_size = 128 #MB 1/4 of max video size(512)
photo_avg_size = 1.25 #MB 1/4 of max photo size(5)
video_cnt = 0
photo_cnt = 0
mb_to_b = 1000000
for df in dfs:
    temp = df['photos']
    photo_cnt += temp.apply(lambda x: x.count('media')).sum()
    video_cnt += temp.apply(lambda x: x.count('video')).sum()
print(photo_cnt, video_cnt)

totoal_size = video_avg_size * video_cnt * mb_to_b + photo_avg_size * photo_cnt * mb_to_b
avg_media_size = totoal_size / tweets_num
other = (i64_count * 4 + f64_count * 4 + strlen_avg)
total_avg_size = avg_media_size + other
```

```python
print(total_avg_size)
print(round(total_avg_size))
```

# Q4

```python
transfer_avg = views_avg * total_avg_size
```

```python
print(transfer_avg)
print(round(transfer_avg))
```
